```python
def min_max_normalize(data):
    min_val = min(data)
    max_val = max(data)
    normalized_data = [(x - min_val) / (max_val - min_val) for x in data]
    return normalized_data
```

```python
llama_7b = min_max_normalize(llama_7b)
llama2_7b = min_max_normalize(llama2_7b)
llama3_8b = min_max_normalize(llama3_8b)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data based on the user's graph
layers = np.arange(0, 33)
llama7b_sensitivity = np.array([200, 400, 350, 500, 1000, 750, 900, 650, 850, 700, 600, 500, 450, 400, 300, 250, 200, 150, 100, 200, 300, 250, 400, 350, 500, 700, 600, 550, 800, 950, 1200, 1000, 900])
llama2_7b_sensitivity = np.array([300, 600, 450, 700, 2000, 1250, 1500, 1150, 1350, 1200, 1100, 1000, 950, 900, 800, 750, 700, 650, 600, 700, 800, 750, 900, 850, 1000, 1200, 1100, 1050, 1300, 1450, 1700, 1500, 1400])
baichuan7b_sensitivity = np.array([250, 500, 400, 600, 1500, 1000, 1200, 950, 1150, 1000, 900, 800, 750, 700, 600, 550, 500, 450, 400, 500, 600, 550, 700, 650, 800, 1000, 900, 850, 1100, 1250, 1500, 1300, 1200])

# Normalize the sensitivity values for bubble sizes
max_sensitivity = max(np.max(llama7b_sensitivity), np.max(llama2_7b_sensitivity), np.max(baichuan7b_sensitivity))
llama7b_bubble_size = (llama7b_sensitivity / max_sensitivity) * 1000  # Scale bubble sizes
llama2_7b_bubble_size = (llama2_7b_sensitivity / max_sensitivity) * 1000
baichuan7b_bubble_size = (baichuan7b_sensitivity / max_sensitivity) * 1000

# Create the bubble plot
plt.figure(figsize=(12, 6))

# Plot bubbles for each model with adjusted positions
plt.scatter(layers, llama7b_sensitivity, s=llama7b_bubble_size, alpha=0.5, color='blue', label='LLaMA-7B')
plt.scatter(layers, llama2_7b_sensitivity, s=llama2_7b_bubble_size, alpha=0.5, color='orange', label='LLaMA2-7B')
plt.scatter(layers, baichuan7b_sensitivity, s=baichuan7b_bubble_size, alpha=0.5, color='green', label='Baichuan-7B')

# Plot lines connecting the tops of the bubbles
plt.plot(layers, llama7b_sensitivity, color='blue')
plt.plot(layers, llama2_7b_sensitivity, color='orange')
plt.plot(layers, baichuan7b_sensitivity, color='green')

plt.xlabel('Layer')
plt.ylabel('Sensitivity')
plt.xticks(layers)
plt.legend()
plt.title('Model Sensitivity by Layer - Bubble Chart')
plt.show()
```

```python
def plot_line_chart(sensitivity):
    # 提取层号和对应的敏感度值
    if isinstance(sensitivity, dict):
        layers = list(sensitivity.keys())
        sensitivity_values = list(sensitivity.values())
    elif isinstance(sensitivity, list):
        layers = list(range(32))  # 创建0到31的层号列表
        sensitivity_values = sensitivity  # 直接使用列表作为敏感度值

    # 创建图形
    plt.figure(figsize=(10, 6))

    # 绘制折线图
    plt.plot(layers, sensitivity_values, label='Sensitivity', color='blue')

    # 添加标签和标题
    plt.xlabel('Layer')
    plt.ylabel('Sensitivity')
    plt.title('Layer-wise Sensitivity')

    # 显示图例
    plt.legend()

    # 显示图形
    plt.show()
```
