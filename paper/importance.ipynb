# 计算重要性

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Example data for three models (You should replace these with your actual data)
layers_7B = np.arange(1, 33)  # Assuming LLaMA-7B has 32 layers
sparsity_7B_uniform = np.ones_like(layers_7B) * 0.7  # Uniform sparsity ratio
sparsity_7B_owl = np.random.normal(0.7, 0.1, size=layers_7B.shape)  # Example OWL data

layers_13B = np.arange(1, 41)  # Assuming LLaMA-13B has 40 layers
sparsity_13B_uniform = np.ones_like(layers_13B) * 0.7  # Uniform sparsity ratio
sparsity_13B_owl = np.random.normal(0.7, 0.1, size=layers_13B.shape)  # Example OWL data

layers_30B = np.arange(1, 61)  # Assuming LLaMA-30B has 60 layers
sparsity_30B_uniform = np.ones_like(layers_30B) * 0.7  # Uniform sparsity ratio
sparsity_30B_owl = np.random.normal(0.7, 0.1, size=layers_30B.shape)  # Example OWL data

# Plotting the graphs
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# LLaMA-7B
axes[0].bar(layers_7B, sparsity_7B_owl, color='lightgreen', alpha=0.5)
axes[0].plot(layers_7B, sparsity_7B_owl, color='orange', linewidth=2, label="OWL (Ours)")  # OWL curve
axes[0].plot(layers_7B, sparsity_7B_uniform, 'k-.', linewidth=2, label="Uniform")
axes[0].set_title('LLaMA-7B')
axes[0].set_xlabel('Layer Index')
axes[0].set_ylabel('Layerwise Sparsity Ratio')
axes[0].legend()

# LLaMA-13B
axes[1].bar(layers_13B, sparsity_13B_owl, color='lightblue', alpha=0.5)
axes[1].plot(layers_13B, sparsity_13B_owl, color='orange', linewidth=2, label="OWL (Ours)")  # OWL curve
axes[1].plot(layers_13B, sparsity_13B_uniform, 'k-.', linewidth=2, label="Uniform")
axes[1].set_title('LLaMA-13B')
axes[1].set_xlabel('Layer Index')
axes[1].set_ylabel('Layerwise Sparsity Ratio')
axes[1].legend()

# LLaMA-30B
axes[2].bar(layers_30B, sparsity_30B_owl, color='lightcoral', alpha=0.5)
axes[2].plot(layers_30B, sparsity_30B_owl, color='orange', linewidth=2, label="OWL (Ours)")  # OWL curve
axes[2].plot(layers_30B, sparsity_30B_uniform, 'k-.', linewidth=2, label="Uniform")
axes[2].set_title('LLaMA-30B')
axes[2].set_xlabel('Layer Index')
axes[2].set_ylabel('Layerwise Sparsity Ratio')
axes[2].legend()

plt.tight_layout()
plt.show()
```
